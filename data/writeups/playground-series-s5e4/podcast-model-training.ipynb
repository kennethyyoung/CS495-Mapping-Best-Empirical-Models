```python
import numpy as np
import pandas as pd
import lightgbm as lgb

import warnings
warnings.simplefilter('ignore')

cat_columns = ['Podcast_Name', 'Publication_Day', 'Publication_Time', 'Episode_Sentiment', 'Episode_Num']
```

```python
df_train = pd.read_parquet('/kaggle/input/podcast-listening-huge-dataset/train.parquet')
df_train[cat_columns] = df_train[cat_columns].astype('category')

df_train.info()
```

```python
y_train = df_train['Listening_Time_minutes']
df_train.drop(columns=['Listening_Time_minutes'], inplace=True)
X_train = df_train
```

```python
model = lgb.LGBMRegressor(
    objective='l2',
    metric='rmse', 
    random_state=42,
    n_iter=12000,
    max_depth=15,
    learning_rate=0.008,
    num_leaves=480,
    colsample_bytree=0.25,
    verbosity=-1,
).fit(X_train, y_train)
```

```python
X_test = pd.read_parquet('/kaggle/input/podcast-listening-huge-dataset/test.parquet')
X_test[cat_columns] = X_test[cat_columns].astype('category')
```

```python
df_subm = pd.read_csv('/kaggle/input/playground-series-s5e4/sample_submission.csv', index_col='id')
df_subm['Listening_Time_minutes'] = model.predict(X_test)
df_subm.to_csv('submission.csv')
```
