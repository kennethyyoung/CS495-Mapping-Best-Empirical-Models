```python
train = pd.read_csv("/kaggle/input/competitions/playground-series-s6e2/train.csv")
test  = pd.read_csv("/kaggle/input/competitions/playground-series-s6e2/test.csv")
orig  = pd.read_csv("/kaggle/input/datasets/cdeotte/s6e4-original-dataset/Heart_Disease_Prediction.csv")
submission = pd.read_csv("/kaggle/input/competitions/playground-series-s6e2/sample_submission.csv")

# map targets
train["Heart Disease"] = train["Heart Disease"].map({"Absence": 0, "Presence": 1}).astype(int)
orig["Heart Disease"]  = orig["Heart Disease"].map({"Absence": 0, "Presence": 1}).astype(int)

# X/y from kaggle train
X = train.drop(columns=["Heart Disease", "id"], errors="ignore")
y = train["Heart Disease"].astype(int)

# keep orig fully labeled; make X_orig/y_orig separately
X_orig = orig.drop(columns=["Heart Disease"], errors="ignore")
y_orig = orig["Heart Disease"].astype(int)
```

```python
CATS = ['Age', 'Sex', 'Chest pain type', 'FBS over 120', 'Exercise angina', 'Thallium']
NUMS = ['BP', 'Cholesterol', 'Max HR', 'ST depression', 'Slope of ST', 'Number of vessels fluro', 'EKG results']

NEW_NUMS, NEW_CATS, NUM_AS_CAT, TO_REMOVE = [], [], [], []

# Frequency encoding from ALL data (train + orig + test)
for cat in NUMS:
    freq = pd.concat([train[cat], orig[cat], test[cat]]).value_counts(normalize=True)
    for df in [train, test, orig]:
        df[f'FREQ_{cat}'] = df[cat].map(freq).fillna(0).astype('float32')
    NEW_NUMS.append(f'FREQ_{cat}')

# Numerical as categorical
for col in NUMS:
    new_col = f'CAT_{col}'
    NUM_AS_CAT.append(new_col)
    for df in [train, test, orig]:
        df[new_col] = df[col].astype(str).astype('category')

FEATURES = NUMS + CATS + NEW_NUMS + NEW_CATS + NUM_AS_CAT

STATS = ['mean']
TE_COLUMNS = NUM_AS_CAT + CATS + NEW_CATS
TO_REMOVE += NUM_AS_CAT + CATS + NEW_CATS
```

```python
# 1) Frequency encoding (use train+orig+test to compute frequencies)
for col in NUMS:
    freq = pd.concat([train[col], orig[col], test[col]]).value_counts(normalize=True)
    orig[f"FREQ_{col}"] = orig[col].map(freq).fillna(0).astype("float32")

# 2) Numeric-as-categorical
for col in NUMS:
    orig[f"CAT_{col}"] = orig[col].astype(str).astype("category")

```

```python
## Safe feature growth ##
corr_scores = {}
for a, b in combinations(X.columns, 2):
    corr_scores[(a, b)] = abs(np.corrcoef(X[a], X[b])[0, 1])

top_pairs = sorted(corr_scores, key = corr_scores.get, reverse=True)[:8]

for a, b in top_pairs:
    X[f"{a}_x_{b}"] = X[a] * X[b]
    test[f"{a}_x_{b}"] = test[a] * test[b]
    orig[f"{a}_x_{b}"] = test[a] * test[b]
    
    
```

```python
import numpy as np
import pandas as pd
import gc
import xgboost as xgb

from catboost import CatBoostClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score

# ====================================================
# SETTINGS
# ====================================================

FOLDS = 15
INNER_FOLDS = 15

# ---------- XGB PARAMS (GPU) ----------
best_params = {
    "n_estimators": 2014,
    "learning_rate": 0.04174204966805775,
    "max_depth": 3,
    "min_child_weight": 3,
    "subsample": 0.8088944727974543,
    "colsample_bytree": 0.5763444919186352,
    "reg_lambda": 0.0003090977633221219,
    "reg_alpha": 0.6664753071198798,

    "objective": "binary:logistic",
    "eval_metric": "auc",
    # "tree_method": "gpu_hist",
    # "predictor": "gpu_predictor",
    # "enable_categorical": True,
    "random_state": 42,
}

# ---------- CATBOOST PARAMS (GPU) ----------
cb_params = {
    "iterations": 8000,
    "learning_rate": 0.01,
    "depth": 5,
    "loss_function": "Logloss",
    "eval_metric": "AUC",
    "auto_class_weights": "Balanced",
    "bootstrap_type": "Bernoulli",
    "subsample": 0.9,
    "l2_leaf_reg": 12,
    "random_strength": 1.2,
    "boosting_type": "Ordered",
    # "task_type": "GPU",
    "devices": "0",
    "verbose": False
}

# ====================================================
# HILLCLIMB BLEND
# ====================================================

def hillclimb_weight(y_val, p1, p2):
    best_w = 0.5
    best_auc = -1

    for w in np.linspace(0, 1, 51):
        blend = w*p1 + (1-w)*p2
        auc = roc_auc_score(y_val, blend)
        if auc > best_auc:
            best_auc = auc
            best_w = w

    return best_w, best_auc


# ====================================================
# CV SETUP
# ====================================================

kf = KFold(n_splits=FOLDS, shuffle=True, random_state=42)

oof = np.zeros(len(train))
pred = np.zeros(len(test))
roc_auc_folds = []

# Orig dataset
X_orig = orig[FEATURES + ['Heart Disease']].reset_index(drop=True).copy()
y_orig = orig['Heart Disease'].reset_index(drop=True).astype(int)

# ====================================================
# OUTER CV LOOP
# ====================================================

for i, (train_index, val_index) in enumerate(kf.split(train)):

    print(f"\nFold {i+1}/{FOLDS}")

    # ---------------------------------------
    # OUTER SPLIT
    # ---------------------------------------
    X_train = train.loc[train_index, FEATURES + ['Heart Disease']].reset_index(drop=True).copy()
    y_train = train.loc[train_index, 'Heart Disease'].reset_index(drop=True).astype(int)

    # Add original dataset
    X_train = pd.concat([X_train, X_orig]).reset_index(drop=True)
    y_train = pd.concat([y_train, y_orig]).reset_index(drop=True)

    X_val = train.loc[val_index, FEATURES].reset_index(drop=True).copy()
    y_val = train.loc[val_index, 'Heart Disease'].values

    X_test = test[FEATURES].reset_index(drop=True).copy()

    # ====================================================
    # INNER KFOLD TARGET ENCODING
    # ====================================================

    kf2 = KFold(n_splits=INNER_FOLDS, shuffle=True, random_state=42)

    # create TE columns
    for col in TE_COLUMNS:
        for s in STATS:
            X_train[f"TE1_{col}_{s}"] = np.nan

    for train_index2, val_index2 in kf2.split(X_train):

        X_train2 = X_train.loc[train_index2, FEATURES + ['Heart Disease']]
        X_val2   = X_train.loc[val_index2, FEATURES]

        for col in TE_COLUMNS:

            tmp = X_train2.groupby(col)['Heart Disease'].agg(STATS)
            tmp.columns = [f"TE1_{col}_{s}" for s in STATS]
            tmp = tmp.astype("float32")

            X_val2 = X_val2.merge(tmp, on=col, how="left")

            for c in tmp.columns:
                X_train.loc[val_index2, c] = X_val2[c].values

    # fill NA TE values
    global_mean = float(np.mean(y_train))
    te_cols_created = [f"TE1_{c}_{s}" for c in TE_COLUMNS for s in STATS]

    for c in te_cols_created:
        X_train[c] = X_train[c].fillna(global_mean).astype("float32")

    # ---------------------------------------
    # FULL TRAIN TE -> VAL + TEST
    # ---------------------------------------
    for col in TE_COLUMNS:

        tmp = X_train.groupby(col)['Heart Disease'].agg(STATS)
        tmp.columns = [f"TE1_{col}_{s}" for s in STATS]

        X_val = X_val.merge(tmp, on=col, how="left")
        X_test = X_test.merge(tmp, on=col, how="left")

    for c in te_cols_created:
        X_val[c] = X_val[c].fillna(global_mean)
        X_test[c] = X_test[c].fillna(global_mean)

    # ---------------------------------------
    # CATEGORICAL TYPES
    # ---------------------------------------
    cat_for_xgb = CATS + NEW_CATS + NUM_AS_CAT

    X_train[cat_for_xgb] = X_train[cat_for_xgb].astype(str).astype("category")
    X_val[cat_for_xgb] = X_val[cat_for_xgb].astype(str).astype("category")
    X_test[cat_for_xgb] = X_test[cat_for_xgb].astype(str).astype("category")

    # remove TE base cols
    X_train.drop(TO_REMOVE, axis=1, inplace=True, errors="ignore")
    X_val.drop(TO_REMOVE, axis=1, inplace=True, errors="ignore")
    X_test.drop(TO_REMOVE, axis=1, inplace=True, errors="ignore")

    # ---------------------------------------
    # FINAL MATRICES
    # ---------------------------------------
    y_train_final = y_train
    X_train_final = X_train.drop(['Heart Disease'], axis=1)

    COLS = X_train_final.columns
    X_val_final = X_val[COLS]
    X_test_final = X_test[COLS]

    # ====================================================
    # XGBOOST
    # ====================================================
    xgb_model = xgb.XGBClassifier(**best_params)

    xgb_model.fit(
        X_train_final, y_train_final,
        eval_set=[(X_val_final, y_val)],
        verbose=False
    )

    val_xgb = xgb_model.predict_proba(X_val_final)[:,1]
    te_xgb  = xgb_model.predict_proba(X_test_final)[:,1]

    # ====================================================
    # CATBOOST
    # ====================================================
    Xtr_cb = X_train_final.copy()
    Xva_cb = X_val_final.copy()
    Xte_cb = X_test_final.copy()

    for c in Xtr_cb.columns:
        if str(Xtr_cb[c].dtype) == "category":
            Xtr_cb[c] = Xtr_cb[c].astype(str)
            Xva_cb[c] = Xva_cb[c].astype(str)
            Xte_cb[c] = Xte_cb[c].astype(str)

    cb_model = CatBoostClassifier(**cb_params, random_seed=42+i)

    cb_model.fit(
        Xtr_cb, y_train_final,
        eval_set=(Xva_cb, y_val),
        use_best_model=True,
        verbose=False
    )

    val_cb = cb_model.predict_proba(Xva_cb)[:,1]
    te_cb = cb_model.predict_proba(Xte_cb)[:,1]

    # ====================================================
    # HILLCLIMB BLEND
    # ====================================================
    w, best_auc = hillclimb_weight(y_val, val_xgb, val_cb)

    val_blend = w*val_xgb + (1-w)*val_cb
    te_blend  = w*te_xgb + (1-w)*te_cb

    oof[val_index] = val_blend
    pred += te_blend

    fold_auc = roc_auc_score(y_val, val_blend)
    roc_auc_folds.append(fold_auc)

    print(f"Blend AUC: {fold_auc:.5f} | best w(XGB)={w:.2f}")

    # cleanup
    del xgb_model, cb_model
    del X_train, X_val, X_test
    del X_train_final, X_val_final, X_test_final
    gc.collect()

# ====================================================
# FINAL RESULTS
# ====================================================

pred /= FOLDS

print("\nOOF AUC:", roc_auc_score(train['Heart Disease'], oof))
print("Fold AUCs:", [round(x,6) for x in roc_auc_folds])
```
