# **DATA STORE CREATION**

```python
%%time 
import pandas as pd
import numpy as np, os

from gc import collect;
from tqdm.notebook import tqdm
```

```python
%%time 

def TransferData(version_nb: int):
    "This function loads the train-test tables from the input kernel outputs and saves them appropriately"
    
    path = f"/kaggle/input/playgrounds4e07-features-v{version_nb}"
    all_files = os.listdir(path)
    
    try:
        os.mkdir(f"/kaggle/working/V{version_nb}")
    except:
        pass
    
    print(f"{'=' * 20} NOTEBOOK VERSION {version_nb} {'=' * 20}")
    pd.read_parquet(os.path.join(path, f"XYtrain_V{version_nb}.parquet")).\
    to_parquet(f"/kaggle/working/V{version_nb}/XYtrain_V{version_nb}.parquet")
    
    pd.read_parquet(os.path.join(path, f"Xtest_V{version_nb}.parquet")).\
    to_parquet(f"/kaggle/working/V{version_nb}/Xtest_V{version_nb}.parquet")
  
```

```python
%%time 

for version_nb in tqdm([2,4,8,12]):
    TransferData(version_nb = version_nb)
    
pd.read_csv(f"/kaggle/input/playground-series-s4e7/sample_submission.csv").to_parquet("sample_submission.parquet")
    
%reset -f
```
